```python
import random
from collections import defaultdict
```

```python
text = """
Artificial intelligence is transforming the world.
Artificial intelligence enables machines to learn.
Machine learning is a subset of artificial intelligence.
Generative models can generate new data.
"""
```

```python
words = text.lower().replace('.', '').split()
```

```python
markov_chain = defaultdict(list)

# Build transition probabilities
for i in range(len(words) - 1):
    current_word = words[i]
    next_word = words[i + 1]
    markov_chain[current_word].append(next_word)
```

```python
def generate_text(chain, start_word, length=20):
    word = start_word
    generated = [word]

    for _ in range(length - 1):
        next_words = chain[word]
        if not next_words:
            break
        word = random.choice(next_words)
        generated.append(word)

    return " ".join(generated)
```

```python
generated_text = generate_text(markov_chain, start_word="artificial", length=30)
print("Generated Text:\n")
print(generated_text)
```
